```python
g = 9.8  # แรงโน้มถ่วง (m/s^2)
h0 = 100  # ความสูงเริ่มต้น (meters)

# กำหนดเวลา t และสร้างสมการการตก
t = sp.symbols('t')
h = h0 - (1/2) * g * t**2

# สร้างฟังก์ชันเชิงตัวเลขจากสมการ
h_func = sp.lambdify(t, h, 'numpy')

# กำหนดช่วงเวลา (0 ถึงเวลาที่ผลไม้ถึงพื้น)
t_vals = np.linspace(0, np.sqrt(2 * h0 / g), 100)

# คำนวณความสูงตามเวลา
h_vals = h_func(t_vals)

# สร้างกราฟ
plt.figure(figsize=(8, 6))
plt.plot(t_vals, h_vals, label='Fruit Agreement', color='b')

# เพิ่มรายละเอียดในกราฟ
plt.title('Graph showing the fall of fruits')
plt.xlabel('time (s)')
plt.ylabel('hight (m)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rain_chances = np.array([20, 15, 10, 5, 25, 40, 50, 45, 30, 20, 15, 10])

plt.figure(figsize=(10, 5))
plt.plot(months, rain_chances, marker='o', color='green')
plt.title('Probability of Rain Each Month')
plt.xlabel('Month')
plt.ylabel('Probability of Rain (%)')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```

```python
years = np.array([2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020])
gold_prices = np.array([19000, 23000, 25000, 20000, 17000, 22000, 24000, 30000, 31000, 35000, 40000])

plt.figure(figsize=(10, 5))
plt.plot(years, gold_prices, marker='o')
plt.title('Gold Prices per Year (Baht)')
plt.xlabel('Year')
plt.ylabel('Price (Baht)')
plt.grid()
plt.xticks(years)
plt.show()
```

```python
years = np.array([2018, 2019, 2020])
boys = np.array([30, 35, 32])
girls = np.array([28, 30, 29])

plt.figure(figsize=(10, 5))
bar_width = 0.35
x = np.arange(len(years))

plt.bar(x, boys, width=bar_width, label='Males', color='lightblue')
plt.bar(x + bar_width, girls, width=bar_width, label='Females', color='lightcoral')
plt.title('Births of Males and Females')
plt.xlabel('Year')
plt.ylabel('Number')
plt.xticks(x + bar_width / 2, years)
plt.legend()
plt.show()
```
